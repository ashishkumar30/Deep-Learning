### Import all lib

```python
import pdfplumber
import pytesseract
from pdf2image import convert_from_path
import fitz
from PIL import Image
```

### Reading a pdf file and converting it to image

```python
filename="Ashish_resume.pdf"
```

### getting text data from pdf file  in string

```python
with pdfplumber.open(filename) as pdf:
    for page in pdf.pages:
        check = page.extract_text()
        print(check)
        break
```

****

### Getting image from pdf file **

### Now getting image part form pdf file 

```python
doc = fitz.open(filename)
for i in range(len(doc)):
    for img in doc.getPageImageList(i):
        xref = img[0]
        pix = fitz.Pixmap(doc, xref)
        if pix.n < 5:  # this is GRAY or RGB
            pix.writePNG("write image")
        #         else:               # CMYK: convert to RGB first
        #             pix1 = fitz.Pixmap(fitz.csRGB, pix)
        #             pix1.writePNG("p%s-%s.png" % (i, xref))
        #             pix1 = None
        #         pix = None
```

```python
Image.open('write image')
```

### Crop image

```python
image_dimensions = Image.open('write image')
# Croping image for itr type
width, height = image_dimensions.size
left = 5
top = height / 20
right = 220
bottom = 3 * height / 6
# Cropped image of above dimension
# (It will not change original image)
croped_image = image_dimensions.crop((left, top, right, bottom))
# Shows the image in image viewer
croped_image
```
